```python
#Q3="""1.Prints 'Palindrome' if it is, otherwise prints 'Not Palindrome',
# print(Q3)
number = input("Enter a number: ")
if number == number[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
for i in range(10):
    count = number.count(str(i))
    if count > 0:
        print(f"{i} appears {count} times")
```
